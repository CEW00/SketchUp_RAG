```python
import os
# Define the directory where the uploaded files are located
txt_files_dir = '/content'  # since you uploaded files directly to this directory

# Read all the text files from the directory
contexts = []
for filename in os.listdir(txt_files_dir):
    if filename.endswith('.txt'):  # make sure it's a .txt file
        with open(os.path.join(txt_files_dir, filename), 'r') as file:
            contexts.append(file.read())  # Read the content of each text file

# Verify that the contents are correctly read
print(contexts[:2])  # Print the first two contexts to check
```

## Question and Answer Generation 

```python
from pprint import pprint
from lmqg import TransformersQG

# initialize model
model = TransformersQG('lmqg/t5-base-squad-qag')
```

## Question and Answer Generation (Multitask)


```python

# Generate question-answer pairs for each context
qa_pairs = []
for context in contexts:
    qa = model.generate_qa(context)  # Generate Q&A for each document
    qa_pairs.append(qa)

# Print the generated Q&A pairs
for i, qa in enumerate(qa_pairs):
    print(f"Q&A for document {i+1}:")
    for question, answer in qa:
        print(f"Q: {question}")
        print(f"A: {answer}")
    print("\n---\n")
```

```python
# Save the Q&A pairs to a text file
output_file = '/content/qa_pairs.txt'
with open(output_file, 'w') as f:
    for i, qa in enumerate(qa_pairs):
        f.write(f"Q&A for document {i+1}:\n")
        for question, answer in qa:
            f.write(f"Q: {question}\nA: {answer}\n")
        f.write("\n---\n")

# Download the file
from google.colab import files
files.download(output_file)
```
